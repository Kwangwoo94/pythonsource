```python
# 1000명의 키와 몸무게를 랜덤하게 파일로 생성

import random

hanguls = list("가나다라마바사아자차카타파하")

with open("data/info.txt","w",encoding="utf-8") as f:
    for i in range(1000):
        name = random.choice(hanguls) + random.choice(hanguls)
        weight = random.randrange(40,100)
        height = random.randrange(140,200)
        f.writelines("{}, {}, {}\n".format(name,weight,height))
```

```python
scores = []

with open("data/score.txt","r",encoding="utf-8") as f:
    
    line = f.readline()
    while line:
        scores.append(int(line))
        line = f.readline()
        
        
print(scores)
print("총합 ",sum(scores))
print("평균 ",sum(scores) / len(scores))
```

### 실습

```python
# score.txt 파일 숫자값을 읽어 총합과 평균을 구한 뒤 result.txt 라는 파일에 결과 작성하기

scores = []
with open("data/score.txt","r",encoding="utf-8") as f:
    for num in f:
        scores.append(int(num))

with open("data/result.txt","w",encoding="utf-8") as f1:
    f1.write("총합 : %d\n" % sum(scores))
    f1.write("평균 : %.2f" % (sum(scores) / len(scores)))
```

```python
# info 파일을 읽어 BMI 지수를 계산한 후 화면 출력하기
# bmi = weight / (height/100) ** 2
# bmi가 25 이상이면 과체중, bmi가 18.5 이상이면 정상체중, 그 아래는 저체중

with open("data/info.txt","r",encoding="utf-8") as f:
    for info in f:
#         print(info.strip())
        name,weight,height = info.strip().split(", ")
    
        bmi = int(weight) / (int(height)/100) ** 2
        
        result = ""
        
        if 25 <= bmi:
            result = "과체중"
        elif 18.5 <= bmi:
            result = "정상체중"
        else:
            result = "저체중"      
   
        print(name,weight,height)
        print("\n".join(["이름 : {}","몸무게:{}", "키:{}", "BMI:{}","결과:{}"])
              .format(name,weight,height,bmi,result))
        print()

# 출력결과
# 차다 43 182
# 이름 : 차다
# 몸무게 : 43
# 키 : 182
# BMI : 12.98125363
# 결과 : 저체중

```

```python
content = ""

while True:
    no = int(input("1.암호화 | 2.복호화 | 3. 종료 중 원하는 메뉴 번호 입력 "))
    
    if no == 1:
        with open("data/origin.txt","r",encoding="utf-8") as f:
            content = f.read()
            
        with open("data/encryption.txt","w",encoding="utf-8") as f:
            for c in content:
                f.write(chr(ord(c)+100))
    elif no == 2:
        with open("data/encryption.txt","r",encoding="utf-8") as f:
            content = f.read()
            for i in range(0,len(content)):
                print(chr(ord(content[i])-100), end="")
    else:
        break
```
